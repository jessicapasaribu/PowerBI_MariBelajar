## Boxplot

```python
import seaborn as sns
```

```python
pip install matplotlib
```

```python
import matplotlib.pyplot as plt
```

```python
features = numerical
for i in range (0, len(features)):
    plt.subplot(1, len(features), i+1)
    sns.boxplot(y=df[features[i]], color="green", orient='v')
    plt.tight_layout()
```

```python
feature = numerical
for i in range(0, len(feature)):
    plt.subplot(2,4, i+1)
    sns.kdeplot(x=df[feature[i]], color='green')
    plt.xlabel(feature[i])
    plt.tight_layout()
```

## Violin

```python
featuress= numerical
for i in range (0, len(featuress)):
    plt.subplot(1,len(featuress), i+1)
    sns.violinplot(y=df[featuress[i]], color='gray', orient='v')
    plt.tight_layout()
```

### countplot

```python
features = categorical
for i in range (0,len(features)):
                plt.subplot(1, len(features), i+1)
                sns.countplot(x=df[features[i]], color='green')
                plt.tight_layout
```

```python
# correlation heatmap
plt.figure(figsize=(8, 8))
sns.heatmap(df[numerical].corr(), cmap='Blues', annot=True, fmt='.2f')
```
